```python
import scipy.io as sio
```

```python
mat_file_path = '/export/mkrishn9/hrtf_field/anthro.mat'
mat_file_path = '/export/mkrishn9/hrtf_field/cipic/anthro.mat'
```

```python
try:
    mat_contents = sio.loadmat(mat_file_path)
    print("File loaded successfully!")
except FileNotFoundError:
    print(f"Error: The file '{mat_file_path}' was not found.")
except Exception as e:
    print(f"An error occurred while loading the .mat file: {e}")
```

```python
print("\nVariables in the .mat file:")
for key, value in mat_contents.items():
    # Ignore '__header__', '__version__', '__globals__' which are usually metadata
    if not key.startswith('__'):
        print(f"- {key}: Type={type(value)}, Shape={value.shape if hasattr(value, 'shape') else 'N/A'}")
```

```python
if 'id' in mat_contents:
    your_data_variable = mat_contents['id']
    print("\nContent of 'data' variable:")
    print(your_data_variable)

    # To see just a few entries, you can slice or look at the first few rows/elements
    print("\nFirst 5 entries (or fewer if less than 5):")
    if your_data_variable.ndim == 1: # 1D array
        print(your_data_variable[:5])
    elif your_data_variable.ndim >= 2: # 2D or higher dimensional array
        print(your_data_variable[0:5, :]) # First 5 rows, all columns
    else:
        print(your_data_variable) # Print all if not an array
else:
    print("\n'data' variable not found. Please check the available keys above.")
```
